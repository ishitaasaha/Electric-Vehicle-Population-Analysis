# *check for outliers*



*   IQR (Interquartile Range)
*    Z-Score



**IQR (Interquartile Range)**

```python
# Convert 'Postal Code' and 'Model Year' columns to numeric types
df['Postal Code '] = pd.to_numeric(df['Postal Code'], errors='coerce')
df['Model Year'] = pd.to_numeric(df['Model Year'], errors='coerce')

# Check for missing values after conversion
print("Missing Values after conversion:")
print(df[['Postal Code', 'Model Year']].isnull().sum())

# Proceed with outlier detection
Q1 = df['Postal Code'].quantile(0.25)
Q3 = df['Model Year'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (df['Postal Code'] < lower_bound) | (df['Model Year'] > upper_bound)

# Print or visualize the indices of the outliers
print(df.index[outliers])

# Optional: Remove outliers from the DataFrame
df_no_outliers = df[~outliers]
```

**Z-Score**

```python
from scipy import stats

# Assuming 'Brand' is the column name in your DataFrame
z_scores = stats.zscore(df['Model Year'])
outliers = (z_scores > 3) | (z_scores < -3)

# Print or visualize the indices of the outliers
print(df.index[outliers])

# Optional: Remove outliers from the DataFrame
df_no_outliers = df[~outliers]
```

```python
# This function 'Model_analysis' classifies vehicle models based on the provided values.
# If the value is less than or equal to 2017, it returns "PACIFICA", if less than or equal to 2012, it returns "LEAF", else "Another Model".
def Model_analysis(values):
    if values <= 2017:
        return "PACIFICA"
    elif values <= 2012:
        return "LEAF"
    else:
        return "Another Model"
```

```python
#Creating a new column 'Model_Year_Analysis' by applying the 'Model_analysis' function to the 'Model Year' column, allowing for insightful analysis and transformation of model years in the DataFrame."
df['Model_Year_Analysis'] =df['Model Year'].apply(Model_analysis)
```

# **T test - mean (Small dataset)**

A t-test is a statistical test used to determine if there is a significant difference between the means of two groups. When dealing with a small dataset, the t-test is often preferred over other tests, such as the z-test, because it accounts for the uncertainty introduced by the small sample size.

```python
# Import the t-test function from the scipy.stats module
from scipy import stats

# Perform an independent two-sample t-test for 'Vehicle Location' based on 'DOL Vehicle ID' values
t_statistics, p_value = stats.ttest_ind(df[df['DOL Vehicle ID'] == 1]['Vehicle Location'].dropna(), df[df['DOL Vehicle ID'] == 0]['Vehicle Location'].dropna())

# Print the calculated t-statistics
print(t_statistics)

# Print the corresponding p-value
print(p_value)
 # null hypothesis will be accepted
```

# **Z test - mean (Large dataset)**
A Z-test for the mean is a statistical test that helps determine whether the mean of a sample is significantly different from a known or hypothesized population mean. It is often used when the sample size is large, and the data are assumed to be approximately normally distributed. Here are some reasons why a Z-test for the mean might be used in the context of a large dataset

```python
# Import the required function from statsmodels
from statsmodels.stats.weightstats import ztest

# Perform a Z-test on the 'Electric Utility' column
z_statistic, p_value = ztest(df['City'].dropna(), value=32)

# Print the results of the Z-test
print("\nZ-Test:")
print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)
```

# **F test - variance (comparing varaince of two samples)**

The F-test for comparing variances is used to determine whether the variances of two independent samples are significantly different. This test is commonly employed in statistical analysis to make inferences about the equality of variances between two groups.

```python
from scipy.stats import f_oneway

# Assuming 'City' is a categorical variable and you want to compare means across different cities
f_statistic, p_value = f_oneway(df['Electric Range'], df['City'])

# Print the results of the ANOVA test
print("\nOne-way ANOVA:")
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
```

# **Anova test - varaince (Variables are categorical)**

# **Kolmogorov-Smirnov (KS)** **test**
The Kolmogorov-Smirnov (KS) test is a non-parametric statistical test that is used to compare the distribution of a sample with a reference probability distribution or to compare the distributions of two independent samples. It's particularly useful when the underlying distributions are not assumed to follow a specific parametric form.

```python
ks_stats,pvalue=stats.kstest(df['Model'],'norm')
print(pvalue) #null hypothesis is acceptedm the column does not exhibit normal distribution
```

# **Man whiteny test**
The Mann-Whitney U test, also known as the Wilcoxon rank-sum test, is a non-parametric test used to assess whether there is a difference between the distributions of two independent samples. It does not assume that the data are normally distributed.

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(df['Model'], df['Model Year'])

# Print the results
print("Mann-Whitney U Statistic:", statistic)
print("P-Value:", p_value)

# Check the p-value against the significance level (e.g., 0.05)
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the groups.")
```
